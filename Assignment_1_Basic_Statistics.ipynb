```python
sns.boxplot(cars.MPG)
```

```python
# P(MPG>38)
1-stats.norm.cdf(38,cars.MPG.mean(),cars.MPG.std())
```

```python
# P(MPG<40)
stats.norm.cdf(40,cars.MPG.mean(),cars.MPG.std())
```
