# **Exercise 1 (Cutlets data exercise)**

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

 **Assume**,

 Null hypothesis h0 = Diameters of the cutlets of the two units are similar

 Alternate hypothesis ha = Diameters of the cutlets of the two units are dissimilar

```python
from google.colab import files
uploaded = files.upload()
```

```python
data=pd.read_csv("Cutlets.csv")
data.head()
```

```python
#let mu1 be the mean of cutlets of unit A and mu2 be the mean of cutlets of unit B
#Hence h0 : mu1 - mu2 = 0 & ha: mu1 - mu2 != 0
unit_A = data['Unit A']
unit_B = data['Unit B']
```

```python
#Since this is a sample we have to use t test
stats.ttest_ind(unit_A , unit_B)
```

**Conclusion**:Here significence level is 5% so alpha is 0.05

p value ~ 0.4722 which is greater than alpha which means that the diameters of the cutlets of the two units are not significantly different

If p value is not equal to  α we failed to reject Null Hypothesis because of lack of evidence

# **Exercise 2 (LabTAT data exercise)**

Here we have 4 different types laboratory so let their respective means be mu1,mu2,mu3,mu4

**Assume**,
Null hypothesis h0 = all the means are equal i.e mu1=mu2=mu3=mu4.

Alternative hypothesis ha = atleast one of the mean is different.

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv("LabTAT.csv")
df.head()
```

```python
df.shape
```

```python
#Since we have to find if the means of 4 columns are equal or not, we have to use f test or anova test
stats.f_oneway(df.iloc[:,0], df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
```

**Conclusion:** Here p value is almost nearly zero and is thus less than alpha or significance level i.e 5% or 0.05.
This means that we reject h0 and accept ha.

Hence, there is indeed a difference in the average TAT values between the 4 laboratories

# **Exercise 3 (BuyerRatio data exercise)**

```python
from google.colab import files
uploaded = files.upload()
```

```python
df1 = pd.read_csv('BuyerRatio.csv')
df1.head()
```

**Assume**, h0 = Proportions are equal and ha = Proportions are not equal

```python
Data = df1.iloc[:,1:5]
```

```python
Data
```

```python
#Since there are counts of males and females across different regions, we use chi square to find relation between them.
```

```python
stats.chi2_contingency(Data)
```

**Conclusion:** Here the p value is 0.6603 which is greater than 0.05 i.e alpha (significance level). Hence proportions are not equal.

This means that we reject h0 and accept ha.
